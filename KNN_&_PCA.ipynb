#Theoretical

1. What is K-Nearest Neighbors (KNN) and how does it work?
ANS: K-Nearest Neighbors (KNN) is a simple yet powerful machine learning algorithm used for classification and regression tasks. It works by comparing new data points to existing ones and making predictions based on similarity.

How does it work:
- KNN doesn't have a training phase like other algorithms. Instead, it simply stores the dataset and makes predictions when needed.
- When given a new data point, KNN calculates the distance between this point and all points in the dataset. The most common distance metrics are:
Euclidean Distance (most widely used)
Manhattan Distance
- The algorithm selects the K closest points (neighbors) based on the calculated distances.
- Making Predictions:
Classification: The algorithm assigns the most common class among the K neighbors to the new data point.
Regression: It takes the average (or weighted average) of the values of the K neighbors.
- Final Decision: The algorithm outputs the predicted label or value for the new data point.



2. What is the difference between KNN Classification and KNN Regression?
ANS: KNN Classification:
- Prpose is Used to classify data into discrete categories (e.g., "Spam" or "Not Spam")
- The output should be a class label.
- The most common class among K-nearest neighbors.
- Example : Identifying whether an email is spam or not

KNN Regression:
- Prpose is Used to predict continuous values (e.g., house prices, temperatures)
- The output should be a numerical value
- The average or weighted average of K-nearest neighbors’ values
- Example: Predicting the price of a house based on features like size and location.

3. What is the role of the distance metric in KNN?
ANS: The distance metric is at the heart of how K-Nearest Neighbors (KNN) determines similarities between data points. It directly impacts how neighbors are selected and ultimately influences classification or regression predictions

Role of the Distance Metric in KNN:
- It measures how close or far a new data point is from existing ones.
- KNN picks the K closest points based on this metric.
- A good distance metric ensures meaningful comparisons, improving prediction quality.
- Some metrics work better for numerical data, others for categorical or text-based inputs.

4. What is the Curse of Dimensionality in KNN?
ANS: The Curse of Dimensionality refers to the challenges that arise when working with high-dimensional data, especially in algorithms like K-Nearest Neighbors (KNN). As the number of features (dimensions) increases, KNN's performance can degrade significantly.


5. How can we choose the best value of K in KNN?
ANS: Choosing the best value of K in K-Nearest Neighbors (KNN) is crucial for achieving optimal model performance. If K is too small or too large, it can negatively impact accuracy.
How to Choose the Best K?
1) Avoiding Too Small K (Overfitting Risk)
- A small K (e.g., K = 1) makes the model highly sensitive to noise.
- It memorizes the training data instead of generalizing, leading to poor results on new data.

2) Avoiding Too Large K (Underfitting Risk)
- A large K (e.g., K = 50) makes predictions too generalized.
- The model may not capture local trends and patterns in the dataset.

3) Using the "Square Root Rule" (Basic Estimate)
- A common heuristic suggests setting: [ K = \sqrt{N} ] where N is the total number of data points.
- This is a simple starting point but should be refined with testing.

4) Using Cross-Validation
- Perform K-Fold Cross-Validation to evaluate different values of K.
- Choose the K that results in the lowest error rate or highest accuracy.

5) Plotting Accuracy vs. K
- Try multiple values of K and plot them against model accuracy.
- Look for the "elbow point" where accuracy stabilizes.

6. What are KD Tree and Ball Tree in KNN?
ANS: KD Tree and Ball Tree are two different data structures used to optimize the K-Nearest Neighbors (KNN) algorithm, especially when dealing with high-dimensional data. They help improve the efficiency of searching for the nearest neighbors by reducing the computational complexity.

KD Tree (K-Dimensional Tree):
- A binary space-partitioning tree that recursively divides the space into regions using axis-aligned splits.
- It best for Low to moderate dimensional data (typically up to 30 dimensions).
- It partitions the data into hierarchical nodes, allowing quick retrieval of nearest points by eliminating large portions of the search space.
- Performance declines as the number of dimensions increases due to the "curse of dimensionality.

Ball Tree:
- A tree structure where each node is a hypersphere (ball) that encapsulates a subset of the points.
- It best for High-dimensional data where KD Tree struggles.
- The tree organizes points into nested spherical clusters, making distance-based queries more efficient in high-dimensional spaces.
- It Works better with arbitrary distance metrics, not just Euclidean distance.

7. When should you use KD Tree vs. Ball Tree?
ANS: Use KD Tree When:
- Your data has low to moderate dimensions (typically up to 30).
- Your working with Euclidean distance or similar structured distance metrics.
- Your dataset is well-distributed and doesn’t contain many overlapping clusters.
- You need faster exact nearest neighbor searches in lower dimensions.

Use Ball Tree When:
- Your data is high-dimensional (above 30 dimensions) or doesn't follow a simple spatial structure.
- You are using arbitrary distance metrics like Manhattan, cosine similarity, or Mahalanobis distance, which KD Tree struggles with.
-  Your dataset has dense clusters, making KD Tree inefficient.
- You need better performance for approximate searches in high-dimensional spaces.

9. How does feature scaling affect KNN?
ANS: Feature scaling plays a crucial role in K-Nearest Neighbors (KNN) because KNN relies on distance measurements to determine the nearest neighbors. If features have vastly different scales, those with larger numerical values can dominate the distance calculations, leading to biased or inaccurate predictions.
Effect of Feature Scaling on KNN:
- If one feature has values in the range of 0-1, while another has values in the range of 0-1000, the second feature will have a much bigger impact on the distance metric. Scaling ensures that all features contribute equally to the similarity calculations.
- Without scaling, KNN may incorrectly classify points because distance measurements won't be fair across different features. Scaling ensures that features with smaller ranges still influence neighbor selection appropriately.
- Distance calculations can be more stable and efficient when features are scaled. It reduces numerical instability and unnecessary complexity in large datasets.
- Common metrics like Euclidean distance and Manhattan distance are highly sensitive to scale differences. Scaling ensures these metrics work consistently across all features.




10. What is PCA (Principal Component Analysis)?
ANS: Principal Component Analysis (PCA) is a dimensionality reduction technique used to: Reduce the number of features in a dataset and While retaining as much variance (information) as possible. It helps transform high-dimensional data into a lower-dimensional form while preserving the most important patterns.

11. How does PCA work?
ANS: Principal Component Analysis (PCA) is a powerful dimensionality reduction technique used in statistics and machine learning. It helps simplify complex datasets by transforming them into a set of new variables called principal components, which capture the most important patterns in the data.
Here's how PCA works:
- Standardization: The data is standardized to ensure that features with different scales don't dominate the analysis.
- Covariance Matrix Computation : A covariance matrix is calculated to understand how variables relate to each other.
- Eigenvalues & Eigenvectors : The eigenvalues and eigenvectors of the covariance matrix are computed. These define the principal components
- Sorting Principal Components : The principal components are ranked based on their explained variance (importance).
- Projection : The data is transformed using the top principal components, reducing dimensions while preserving variance.



12.  What is the geometric intuition behind PCA?
ANS: The geometric intuition behind Principal Component Analysis (PCA) revolves around understanding data in terms of its orientation and variance in a high-dimensional space.
- Data as a Cloud of Points : Imagine your dataset as a cloud of points in an n-dimensional space. Each point represents an observation with multiple features.
- Finding the Best-Fitting Axes : PCA searches for a new set of axes that best describe the spread of data. These axes, called principal components, are chosen to capture the maximum variance.
- Rotating the Coordinate System : Instead of using the original feature axes, PCA finds a new set of orthogonal directions (eigenvectors) that better align with the data's natural shape. These directions are determined from the covariance matrix.
- Projecting the Data : The original data points are projected onto the principal components, which effectively rotates and reorients the dataset to highlight its most significant variation
- Dimensionality Reduction : Since higher principal components contribute less variance, we can ignore them, reducing the number of dimensions while preserving most of the information.



13. What is the difference between Feature Selection and Feature Extraction?
ANS: Feature Selection:
- It Choose the most relevant features from the original dataset while discarding the less important ones.
- It evaluates existing features based on certain criteria (e.g., correlation, mutual information, statistical tests) and removes those that contribute less to prediction accuracy.
- Filter methods (statistical tests), Wrapper methods (testing feature subsets), and Embedded methods (built into models like Lasso regression).
- Example: Keep only height, weight from many features

Feature Extraction:
- It Create new features by transforming or combining existing ones to improve the dataset's representation.
- It generates new feature spaces by applying mathematical transformations such as PCA (Principal Component Analysis), Autoencoders, or t-SNE.
- Linear transformations (PCA), nonlinear transformations (neural networks), or manual combinations of features.
- Example: Combine height & weight into BMI-like feature

14. What are Eigenvalues and Eigenvectors in PCA?
ANS: Eigenvalues and eigenvectors play a crucial role in Principal Component Analysis (PCA) because they help identify the directions in which data varies the most.
Eigenvectors:
- These are the directions along which data is stretched the most.
- In PCA, they define the principal components, which are new axes aligned with the greatest variance in the dataset.
- They are unit vectors that point in these important directions, helping to reorient the data into a new coordinate system.

Eigenvalues:
- These tell us how much variance exists along the eigenvector's direction.
- A larger eigenvalue means more information (variance) is captured along that eigenvector.
- In PCA, we rank principal components based on their eigenvalues—higher eigenvalues mean the component explains more of the dataset’s spread.

15. How do you decide the number of components to keep in PCA?
ANS: In Principal Component Analysis (PCA), choosing the number of components to keep is crucial for balancing dimensionality reduction and retaining enough meaningful information. Here are some common approaches:
1) Explained Variance Ratio: A common rule is to retain components that together explain a significant proportion (e.g., 95%) of the total variance. The cumulative explained variance plot helps visualize this.
2) Kaiser Criterion: If eigenvalues are greater than 1, they indicate that the corresponding principal components contribute more than an individual original feature.
3) Scree Plot: This graph plots the eigenvalues of principal components. The “elbow” point—where the curve starts to flatten—suggests an ideal number of components.
4) Cross-Validation: Evaluating how well different numbers of components perform in a downstream task, like clustering or classification, helps make a more practical decision.
5) Domain Knowledge: Sometimes, choosing a few principal components based on understanding the data and its use case is the best approach.


16.  Can PCA be used for classification?
ANS: PCA itself is not a classification technique, but it can be useful before classification. It transforms high-dimensional data into a lower-dimensional space while preserving variance, which can help improve classification models in several ways
- Feature Reduction: PCA reduces the number of features while retaining most of the important information, making classification models more efficient
- Noise Reduction: Removing less informative components can help models focus on relevant patterns.
- Improved Separability: Sometimes, PCA-transformed features help classifiers distinguish classes better in lower-dimensional space.
- Visualization Aid: PCA allows you to project high-dimensional data onto 2D or 3D for better visualization of class separability.


17. What are the limitations of PCA?
ANS: 1) Loss of Interpretability: PCA transforms original features into principal components, making it harder to interpret the meaning of the transformed features.
2) Linear Assumption: PCA assumes that variability in the data is captured by linear combinations of features, which may not hold true for complex, non-linear relationships.
3) Sensitivity to Scaling: PCA is highly influenced by the scale of features—if features are not standardized, components may be dominated by large-magnitude variables.
4) Ignoring Class Labels: Since PCA is an unsupervised technique, it does not consider class labels, meaning it may not be optimal for classification tasks
5) Poor Performance on Sparse Data: If the dataset has many zero or near-zero values, PCA may struggle to find meaningful components
6) Computationally Expensive: For very large datasets, calculating eigenvectors and eigenvalues can be computationally demanding.


18. How do KNN and PCA complement each other?
ANS: K-Nearest Neighbors (KNN) and Principal Component Analysis (PCA) can complement each other effectively when dealing with high-dimensional data in classification tasks.
- Dimensionality Reduction: PCA reduces the number of features while preserving most of the data's variance, making KNN computationally faster and reducing memory usage.
- Noise Reduction: By eliminating less informative dimensions, PCA helps KNN focus on meaningful features, improving classification accuracy
- Avoiding Curse of Dimensionality: KNN relies on distance metrics, which become less effective in very high-dimensional spaces. PCA mitigates this issue by reducing dimensions.
- Better Visualization: PCA allows data to be projected onto 2D or 3D space, helping to visualize how KNN makes decisions.

However, since PCA is an unsupervised technique, it does not consider class labels, which might sometimes lead to suboptimal transformations for KNN. It’s useful to experiment with different numbers of principal components to find the right balance.

19. How does KNN handle missing values in a dataset?
ANS: K-Nearest Neighbors (KNN) does not inherently handle missing values, but there are strategies to manage them before applying KNN for classification or regression.
- Imputation with Mean, Median, or Mode: Before running KNN, missing values in numerical features can be replaced with the mean or median, while categorical features can use the mode.
- KNN-Based Imputation: Instead of using simple statistical methods, KNN can be used to fill in missing values by predicting them based on their nearest neighbors in the dataset.
- Removing Rows with Missing Values: If missing values are minimal, removing affected rows may be a reasonable option.
- Using a Separate Category for Missing Values: For categorical features, missing values can be assigned a special category like “Unknown” to allow KNN to process them without deletion or imputation.

Among these approaches, KNN-based imputation tends to preserve data structure better than mean or median imputation, especially in cases where missing values are correlated with other features.



20. What are the key differences between PCA and Linear Discriminant Analysis (LDA)?
ANS: PCA:
- It focuses on maximizing variance and finding the most informative directions in the data without considering class labels.
- In PCA, an unsupervised method—it does not require labels.
- It uses eigenvalue decomposition of the covariance matrix to project data onto directions with maximum variance.
- It can retain any number of components, up to the number of original features.
- It used for feature extraction, noise reduction, and visualization.

Linear Discriminant Analysis (LDA):
- Aims to maximize the separability between known classes by finding directions that best distinguish different categories.
- In LDA, a supervised method—it needs class labels for training.
- It uses eigenvalue decomposition of the scatter matrix to find directions that maximize separation between classes.
- The number of components is limited to (number of classes - 1).
- It used primarily in classification tasks to improve class separation.

```python
from sklearn.datasets import load_iris
data = load_iris()
X = data.data [:, :2]
y = data.target

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.3, random_state=42)

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

from sklearn.neighbors import KNeighborsClassifier
from matplotlib.colors import ListedColormap

h = .02
cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF'])
cmap_bold  = ListedColormap(['#FF0000', '#00FF00', '#0000FF'])

def plot_knn(k):
   clf = KNeighborsClassifier(n_neighbors=k)
   clf.fit(X_train_scaled, y_train)

   x_min, x_max = X_train_scaled[:, 0].min() - 1 , X_train_scaled[:, 0].max() + 1
   y_min, y_max = X_train_scaled[:, 1].min() - 1 , X_train_scaled[:, 1].max() + 1
   xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
   Z= clf.predict(np.c_[xx.ravel(), yy.ravel() ])
   Z = Z.reshape(xx.shape)


   plt.figure(figsize=(6, 4))
   plt.contourf(xx, yy, Z, cmap=cmap_light)
   plt.scatter(X_train_scaled[:, 0], X_train_scaled [:, 1], c=y_train, cmap= cmap_bold, edgecolor='k', s =30)
   plt.title(f"KNN Decision Boundary (k = {k})")
   plt.xlabel('Sepal length (scaled)')
   plt.ylabel('Sepal width (scaled)')
   plt.tight_layout()
   plt.show()

for k in [1, 3, 5, 10]:
    plot_knn(k)
```

28. Perform Hyperparameter Tuning on a KNN Classifier using GridSearchCV.

35. Train a KNN Classifier using the KD Tree and Ball Tree algorithms and compare performance.

```python
import time
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

def evaluate_knn(X_train, X_test, y_train, y_test, algorithm):

    knn = KNeighborsClassifier(n_neighbors=5, algorithm=algorithm)

    start_train = time.time()
    knn.fit(X_train, y_train)
    end_train = time.time()

    start_pred = time.time()
    y_pred = knn.predict(X_test)
    end_pred = time.time()


    accuracy = accuracy_score(y_test, y_pred)

    return {
        'Algorithm': algorithm,
        'Accuracy': accuracy,
        'Train Time (s)': end_train - start_train,
        'Predict Time (s)': end_pred - start_pred
    }

def main():

    iris = load_iris()
    X = iris.data
    y = iris.target

    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.3, random_state=42)

    results = []
    for algo in ['kd_tree', 'ball_tree']:
        result = evaluate_knn(X_train, X_test, y_train, y_test, algo)
        results.append(result)

    return results


results = main()

print("{:<12} {:<10} {:<16} {:<16}".format("Algorithm", "Accuracy", "Train Time (s)", "Predict Time (s)"))
for res in results:
    print("{:<12} {:.4f}     {:.6f}         {:.6f}".format(
        res['Algorithm'], res['Accuracy'], res['Train Time (s)'], res['Predict Time (s)']))
```

43.  Train a KNN Classifier and evaluate using ROC-AUC score.

```python
from sklearn.datasets import load_wine

wine = load_wine()
X = wine.data
y = wine.target

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)

from sklearn.metrics import roc_auc_score, roc_curve
from sklearn.preprocessing import label_binarize

y_pred_prob = knn.predict_proba(X_test)

y_test_bin = label_binarize(y_test, classes=[0, 1, 2])

roc_auc = {}
fpr = {}
tpr = {}
for i in range(3):
    fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], y_pred_prob[:, i])
    roc_auc[i] = roc_auc_score(y_test_bin[:, i], y_pred_prob[:, i])

plt.figure(figsize=(10, 6))
for i in range(3):
    plt.plot(fpr[i], tpr[i], label=f'Class {i} (AUC = {roc_auc[i]:.2f})')

plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
plt.title('KNN Classifier ROC Curves for Each Class')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc='lower right')
plt.grid(True)
plt.tight_layout()
plt.show()


for i in range(3):
    print(f"Class {i} - AUC: {roc_auc[i]:.2f}")
```
